### Analysis

Of four drugs identified and analyzed - Capomulin, Ramicane, Infubinol, and Ceftamin - Ramicane appears to be the most effective at minimizing tumor growth with the lowest mean and median tumor size and the longest lived test subjects. This is deducted from looking at the quartile analysis and boxplot diagrams of the four treatments as well as the bar chart analysis of number of timepoints measured. 

Capomulin can be identified as having very similar effects however, and may warrant additional testing. 

 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata, study_results, how = 'outer', on='Mouse ID')

# Display the data table for preview
combined_df.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
tumor_vol_mean_by_drug = clean_df.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
tumor_vol_med_by_drug = clean_df.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
tumor_vol_var_by_drug = clean_df.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
tumor_vol_std_dev_by_drug = clean_df.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
tumor_vol_SEM_by_drug = clean_df.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']

tumor_vol_summary_table = pd.DataFrame({"Mean Tumor Volume": tumor_vol_mean_by_drug,
                                        "Median Tumor Volume": tumor_vol_med_by_drug,
                                        "Tumor Volume Variance": tumor_vol_var_by_drug,
                                        "Tumor Volume Std. Dev.": tumor_vol_std_dev_by_drug,
                                        "Tumor Volume Std. Err.": tumor_vol_SEM_by_drug,
    
})

tumor_vol_summary_table
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
compressed_tumor_vol_summary_table = clean_df.groupby('Drug Regimen').agg({'Tumor Volume (mm3)' : [np.mean, np.median, np.var, np.std, st.sem]})

compressed_tumor_vol_summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
sorted_drugs_df = clean_df_sort.loc[(clean_df_sort['Drug Regimen'] == 'Capomulin') | (clean_df_sort['Drug Regimen'] == 'Ramicane')| (clean_df_sort['Drug Regimen'] =='Infubinol') | (clean_df_sort['Drug Regimen'] =='Ceftamin')]

# Start by getting the last (greatest) timepoint for each mouse
mouse_final_timepoint = sorted_drugs_df.groupby(['Mouse ID'])['Timepoint'].max()

mouse_final_timepoint_df = mouse_final_timepoint.to_frame()
mouse_final_timepoint_df = mouse_final_timepoint_df.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

mouse_end_results_df = pd.merge(mouse_final_timepoint_df, mouse_final_volume_df, how = 'inner', on='Mouse ID')

mouse_end_results_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
drug_regimen_list = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)

drug_final_tumor_vol = []
 
for drug in drug_regimen_list:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    #use iloc func or .loc
    drug_volume = mouse_end_results_df.loc[mouse_end_results_df['Drug Regimen'] == drug, 'Final Tumor Volume (mm3)']
    drug_final_tumor_vol.append(drug_volume)
    
    # Calculate the IQR and quantitatively determine if there are any potential outliers.
    mouse_end_quartiles_df = drug_volume.quantile(q = [.25,.5,.75])
    lowerq = mouse_end_quartiles_df[.25]
    mean = mouse_end_quartiles_df[.5]
    upperq = mouse_end_quartiles_df[.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (iqr*1.5)
    upper_bound = upperq + (iqr*1.5)

    
# add subset 

    print(f'The lower quartile of {drug} is: {lowerq}')
    print(f'The mean of {drug} is: {mean}')
    print(f'The upper quartile of {drug} is: {upperq}')
    print(f'The interquartile range of {drug} is: {iqr}\n')
    print(f'Tumor volumes of {drug} trials below {lower_bound} may be outliers')
    print(f'Tumor volumes of {drug} trials above {upper_bound} may be outliers\n')

    
# Determine outliers using upper and lower bounds
    outliers = drug_volume.loc[(drug_volume<lower_bound)|(drug_volume>upper_bound)]
    
    print(f'The below for {drug} should be examined as outliers:\n')
    if len(outliers) >=1:
        print(f'{outliers}\n')
    else:
        print(f'There are no potential outliers identified for {drug}\n')      
        
    

```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

plt.boxplot(drug_final_tumor_vol, labels = drug_regimen_list)
plt.ylabel('Final Tumor Volume (mm3)')
plt.xlabel('Drug Regimen Used')
plt.title('Final Tumor Volumes By Selected Drug Regimens')
plt.show()
```
